```python
import pandas as pd
```

```python
df = pd.read_csv('Sample - Superstore.csv', index_col=False, encoding='cp1251')

df.info()
```

```python
import matplotlib

%matplotlib inline
```

```python
import datetime
import time
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%m/%d/%Y')
```

```python
df.info()
```

```python
df['ShipYearMonth'] = df['Ship Date'].dt.year * 100 + df['Ship Date'].dt.month
```

```python
df_profit_by_month = df[['ShipYearMonth', 'Profit']].groupby(by=['ShipYearMonth']).sum().reset_index()
```

```python
df_profit_by_month
```

```python
import matplotlib.pyplot as plt
```

```python
#avg_text = "Average refresh time = " + '{:.2f}'.format(avg_time)
title_text = "Profit by Month and Year"
#fig.set_title( title_text  )

base_y = 11
base_x = 2.2 * base_y


plt.figure(figsize=(base_x, base_y))

plt.title(title_text, {'fontsize':24 } )

#plt.line(df_h["Date"], df_h["AvgDuration"])
#plt.figure(figsize=(10,200))


colors = ["royalblue"]

plt.bar(df_profit_by_month['ShipYearMonth'], df_profit_by_month['Profit'], color = colors )


for x,y in zip( df_profit_by_month['ShipYearMonth'], df_profit_by_month['Profit']):

    label = "{:.2f}".format(y)

    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center',  # horizontal alignment can be left, right or center
                fontsize=15   # font size
                )

plt.show()
```
